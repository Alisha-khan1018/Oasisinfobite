# Clean data set

```python
df = df.dropna() #dropping missing values 
```

# Exploring unemployemnt rate by region 

```python
avg_unemployment_rate = df.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()
print(avg_unemployment_rate)
```

```python
print(f"Highest unemploment rate of Indian state is {avg_unemployment_rate.idxmax()} which is {avg_unemployment_rate.max()}")
```

```python
print(f"Lowest unemploment rate of Indian state is {avg_unemployment_rate.idxmin()} which is {avg_unemployment_rate.min()}")
```

# Unemployment trend

```python
plt.figure(figsize=(10, 6))
plt.plot(df[' Estimated Unemployment Rate (%)'])
plt.title('Unemployment Rate Over Time')
plt.xlabel(' Date')
plt.ylabel('Unemployment Rate')
plt.grid()
plt.show()
```

```python
sns.boxplot(df[' Estimated Unemployment Rate (%)'])
plt.title('Boxplot of Unemployment Rate')
plt.show()
```
